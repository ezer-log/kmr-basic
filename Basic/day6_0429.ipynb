## 1.1 가장 간단한 그래프, Line Plot

1.1.1. 첫번째 그래프 : 주어진 데이터가 하나일 때 (plot 함수의 데이터 인자가 1개일 때)

```python
x = np.arange(1, 10)
plt.plot(x)
plt.show()

#이런 경우, x축은 0, 1, 2, 3, ... 으로 자동 지정되고 넣은 데이터는 각 x축에 대응하는 y값이 된다.
```

1.1.2. 두번째 그래프 : x, y축 데이터들이 모두 주어졌을 때

```python
x = np.arange(1, 10)
y = x*5

plt.plot(x, y)
plt.show() #그래프를 보여줘!

# x,y 데이터가 (x, y) 좌표쌍으로 표현되어 그래프에 나타나는 것을 확인할 수 있다.
```

### 1.1.3. 세번째 그래프 : 라인의 컬러 바꾸기

```python
x = np.arange(1, 10)
y = x*5

plt.plot(x, y, 'r') #위의 코드에서 'r'추가
plt.show() #그래프를 보여줘!

# 코드에 'r'을 추가한 것만으로 그래프가 빨간색으로 그려진 것을 확인할 수 있다.
# 'r'은 red의 약자이고, 색을 지정하려면 plot에서 인자를 하나 추가해주면 된다.
```

### 1.1.4. 네번째 그래프 : 라인 마커 변경하기

다양한 선의 종류를 선택할 수 있다. default 값은 직선이고, 점으로 표현하는 마커나 점선 등을 원한다면 선택할 수 있다.


선의 종류는 plot에서 인자로 지정하면 되고, 색을 같이 지정하려면 같은 인자에 함께 문자형으로 색을 지정하면 된다.


예시를 살펴보자! 동그란 마커 ‘o’를 붉은색 ‘r’로 표현하기 때문에, 세번째 인자를 ‘or’로 전달하였다.

```python
x = np.arange(1,10)
y = x*5

plt.plot(x,y,'vr') #여기서 r은 red인데 o가 또 추가되었네! (컬러+마커 같이)
plt.show()
```

```python
plt.plot(x, y, '*g')
plt.show()
```

### 1.1.5. 다섯번째 그래프 : 축의 이름 / 그래프 제목 달기

> 사용 방법:

1. x축 이름 : `plt.xlabel` (x축 이름)
2. y축 이름 : `plt.ylabel` (y축 이름)
3. 그래프 제목 : `plt.title` (그래프 제목)

```python
x = np.arange(1, 10)
y = x*5

plt.plot(x, y)
plt.xlabel('x axis') # x축 이름 지정
plt.ylabel('y axis') # y축 이름 지정
plt.title('matplotlib sample') # 그래프 이름 지정

plt.show()
```

### 1.1.6. 여섯번째 그래프: 구간 설정

x축과 y축의 구간을 설정할 수 있다.

> 사용 방법:

1. x축 구간 설정 : `plt.xlim` (시작 범위, 끝 범위)
2. y축 구간 설정 : `plt.ylim` (시작 범위, 끝 범위)

```python
plt.plot(x, y)
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('matplotlib sample')
plt.xlim(2, 3)
plt.ylim(5, 20)
plt.show()
```

### 1.1.7. 일곱번째 그래프: 범주 달기 (legend)

> 사용 방법:


`plt.legend` (loc = 원하는 위치) #위치 지정해주어야 함

```python
x = np.arange(1,10,0.1)
y1 = x*0.2
y2 = np.sin(x)

plt.plot(x,y1,'b',label='first') #파란색 / first 라벨 부여
plt.plot(x,y2,'r',label='second') #빨간색 / second 라벨 부여
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('matplotlib sample')
plt.legend(loc='upper right') # 범주 표현

plt.show()
```

1.1.8. 여덟번째 그래프: 어노테이션 (화살표로 가리키기)

> 사용 방법:


`plt.annotate (원하는 문자열, xy(화살표 꼭지 xy좌표), xytext(문자열 xy 좌표), arrowprops(화살표 속성 ex.color))`

```python
x = np.arange(1, 10)
y = x*5

plt.plot(x, y, 'b', label = 'total progress')
plt.xlabel('time')
plt.ylabel('achievement')
plt.title('Data Visualization AtoZ')
plt.legend(loc = 'lower right')
plt.annotate('here is our progress', xy = (1.5, 7.5), xytext = (3, 11), arrowprops = {'color': 'green'})
plt.show()
```

### 1.1.9. 아홉번째 그래프: 서브 플롯 그리기

> 사용 방법:

`plt.subplot(nrow,ncol,pos)`

```python
x = np.arange(1,10)
y1 = x*5
y2 = x*1
y3 = x*0.3
y4 = x*0.2

plt.subplot(2,2,1) #2행 2열 (2X2) 그래프 중 (1,1)의 위치
plt.plot(x,y1)

plt.subplot(2,2,2) #2행 2열 (2X2) 그래프 중 (1,2)의 위치
plt.plot(x,y2)

plt.subplot(2,2,3) #2행 2열 (2X2) 그래프 중 (2,1)의 위치
plt.plot(x,y3)

plt.subplot(2,2,4) #2행 2열 (2X2) 그래프 중 (2,2)의 위치
plt.plot(x,y4)

plt.show()
```

그래프가 그려지는 순서는 행방향으로 진행된다는 것을 알 수 있다.

### 1.1.10. 열번째 그래프: 그래프 사이즈

> 사용 방법:

`plt.figure(figsize=(가로,세로))`

```python
x = np.arange(1,10)
y1 = x*5
y2 = x*1
y3 = x*0.3
y4 = x*0.2

plt.figure(figsize=(20,5)) #가로 20, 세로 5로 설정

plt.subplot(2,2,1)
plt.plot(x,y1)

plt.subplot(2,2,2)
plt.plot(x,y2)

plt.subplot(2,2,3)
plt.plot(x,y3)

plt.subplot(2,2,4)
plt.plot(x,y4)

plt.show()
```

figsize를 어떤 숫자로 지정하느냐에 따라 그래프의 크기가 달라진다.

## 3.2. folium을 사용해보자, Advanced

```python
import json
import requests
```

### 3.2.1.1. import data (데이터 불러오기)

```python
url = 'https://raw.githubusercontent.com/python-visualization/folium/master/examples/data'
antarctic_ice_edge = f'{url}/antarctic_ice_edge.json'
antarctic_ice_shelf_topo = f'{url}/antarctic_ice_shelf_topo.json'
```

### 3.2.1.2. 지도 객체 및 레이어 설정 후 지도 그려보기!

1. 지도 객체는 위에서 언급한 `folium.Map`을 통해 지정
2. `folium.GeoJson` 이용하여 GeoJSON 데이터 활용
3. `folium.TopoJson` 이용하여 topoJSON 데이터 활용

```python
# 지도 객체 설정
map_osm = folium.Map(
    location=[-59.1759, -11.6016],
    zoom_start=2
)

# GeoJson 레이어 설정
folium.GeoJson(
    antarctic_ice_edge,
    name='geojson'
).add_to(map_osm)

# TopoJson 레이어 설정
folium.TopoJson(
    json.loads(requests.get(antarctic_ice_shelf_topo).text), #json 모듈 필요
    'objects.antarctic_ice_shelf',
    name='topojson'
).add_to(map_osm)

# 오른쪽 상단에 여러 개의 레이어를 설정할 수 있도록 하는 옵션
folium.LayerControl().add_to(map_osm)

map_osm
```
