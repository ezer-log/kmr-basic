### 배열 생성 및 초기화

* `zeros()`: 모든 요소를 0으로 초기화

```
np.zeros(10)
```

* `ones()`: 모든 요소를 1로 초기화

```
np.ones((3,3))
```

* `full()`: 모든 요소를 지정한 값으로 초기화

```
np.full((3,3), 1.23)
```

* `eye()`: 단위행렬(identity matrix) 생성
  + 주대각선의 원소가 모두 1이고 나머지 원소는 모두 0인 정사각 행렬

```
# 단위행렬 = 정사각형 행렬이라 모양을 정해주는게 아니라 길이(n)만 정해주면 된다
# diagonal = 1, rest = 0
np.eye(3)
```

* `tri()`: 삼각행렬 생성

```
# 삼각행렬도 n만 정해주면 된다
np.tri(3)
```

* `empty()`: 초기화되지 않은 배열 생성
  + 초기화가 없어서 배열 생성비용 저렴하고 빠름
  + 초기화되지 않아서 기존 메모리 위치에 존재하는 값이 있음

```
#초기화 안하다보니 기존 메모리 위치에 있던 값이 불러와진다 = 어떤 값이 올지 모름
np.empty(10)
```

* `_like()`: 지정된 배열과 shape가 같은 행렬 생성
  + `np.zeros_like()`
  + `np.ones_like()`
  + `np.full_like()`
  + `np.empty_like()`

```
print(a1) #위에서 만든 배열
np.zeros_like(a1) #지정 배열과 같은 shape인 배열을 만들어줘
```

```
print(a2)
np.ones_like(a2)
```

```
print(a3)
np.full_like(a3, 10)
```

## 배열 조회

### 배열 속성 정보

```
def array_info(array):
  print(array)
  print("ndim:", array.ndim) #dimension의 갯수. 1이면 1차원
  print("shape:", array.shape)
  print("dtype:", array.dtype) #datatype    
  print("size:", array.size) #number of elements
  print("itemsize:", array.itemsize) #하나의 item의 size, 여기선 1개가 int이니까 8byte
  print("nbytes:", array.nbytes) #size * itemsize = 전체 배열의 bytes 크기
  print("strides:", array.strides) #1차원이니까 하나의 element를 넘어가는데 8byte 필요. 2차원 등에서 하나의 차원에서 다음 차원으로 넘어갈때는 한줄의 total byte 만큼 필요 
```

```
array_info(a1)
```

```
array_info(a2)
```

```
array_info(a3)
```

### 인덱싱(Indexing)

```
print(a1)
print(a1[0])
print(a1[2])
print(a1[-1])
print(a1[-2])
```

```
print(a2)
print(a2[0, 0])
print(a2[0, 2])
print(a2[1, -1])
print(a2[2, -2])
```

```
print(a3)
print(a3[0, 0, 0])
print(a3[1, 1, 1])
print(a3[2, 2, 2])
print(a3[2, -1, -1])
```

### 슬라이싱(Slicing)

* 슬라이싱 구문: `a[start:stop:step]`
* 기본값: start=0, stop=ndim, step=1

```
print(a1)
print(a1[0:2])
print(a1[0:])
print(a1[:1])
print(a1[::2])
print(a1[::-1])
```

```
print(a2)
print(a2[1])
print(a2[1, :])
print(a2[:2, :2])
print(a2[1:, ::-1])
print(a2[::-1, ::-1])
```

### 불리언 인덱싱(Boolean Indexing)

* 배열 각 요소의 선택 여부를 불리언(True or False)로 지정
* True 값인 인덱스의 값만 조회

```
print(a1)
bi = [False, True, True, False, True]
print(a1[bi]) #True 위치에 있는 값들만 표시
```

```
print(a2)
bi = np.random.randint(0, 2, (3, 3), dtype=bool) #bool은 range (0,2) => 0,1 둘중 하나여야 함
print(bi)
print(a2[bi])
```

### 팬시 인덱싱(Fancy Indexing)

```
print(a1)
print([a1[0], a1[2]])

ind = [0, 2] #index를 묶어서 한번에 전달
print(a1[ind]) #해당 index의 값들을 출력

ind = np.array([[0, 1],
                [2, 0]]) #index에 따라서 결과에 대한 shape이 결정됨
print(a1[ind])
```

```
print(a2)
print("---")
row = np.array([0,2])
col = np.array([1,2])
print(a2[row, col])
print("---")
print(a2[row, :])
print(a2[:, col])
print("---")
print(a2[row, 1]) ## Q: 왜 세로로 [2] [8] 이 아닌걸까?
print(a2[row, 0])
print(a2[1, col])
print("---")
print(a2[row, 1:])
print(a2[1:, col])
print("---")
```

## 배열 값 삽입/수정/삭제/복사

### 배열 값 삽입

* `insert()`: 배열의 특정 위치에 값 삽입
* axis를 지정하지 않으면 1차원 배열로 변환
* 추가할 방향을 axis로 지정
* 원본 배열 변경없이 새로운 배열 반환

```
print(a1)
b1 = np.insert(a1, 0, 10) #위치는 0, 값은 10
print(b1)
print(a1) #a1는 값이 바뀌지 않는다
c1 = np.insert(a1, 2, 10)
print(c1)
```

```
print(a2)
b2 = np.insert(a2, 1, 10, axis=0) # "axis=0"=>행. 1번째 행에다 insert
print(b2)
c2 = np.insert(a2, 1, 10, axis=1) # "axis=1"=>열. 1번째 열에다 insert
print(c2)
```

### 배열 값 수정

* 배열의 인덱싱으로 접근하여 값 수정

```
print(a1)
a1[0] = 1
a1[1] = 2
a1[2] = 3
print(a1)

#slicing으로 바꾸기
a1[:1] = 9
print(a1)

#fancy indexing으로 바꾸기
i = np.array([1, 3, 4])
a1[i] = 0
print(a1)
a1[i] += 4
print(a1)
```

```
print(a2)
a2[0, 0] = 1
a2[1, 1] = 2
a2[2, 2] = 3
a2[0] = 1
print(a2)
```

### 배열 연결

* `concatenate()`: 튜플이나 배열의 리스트를 인수로 사용해 배열 연결

* `vstack()`: 수직 스택(vertical stack), 1차원으로 연결

* `hstack()`: 수평 스택(horizontal stack), 2차원으로 연결

* `dstack()`: 깊이 스택(depth stack), 3차원으로 연결

* `stack()`: 새로운 차원으로 연결

### 배열 분할

* `split()`: 배열 분할

```
a1 = np.arange(0, 10)
print(a1)
b1, c1 = np.split(a1, [5])
print(b1, c1)
```

* `vsplit()`: 수직 분할, 1차원으로 분할

```
a2 = np.arange(1, 10).reshape(3, 3)
print(a2)
b2, c2 = np.vsplit(a2, [2])
print(b2)
print(c2)
```

* `hsplit()`: 수평 분할, 2차원으로 분할

```
a2 = np.arange(1, 10).reshape(3, 3)
print(a2)
b2, c2 = np.hsplit(a2, [2])
print(b2)
print(c2)
```

* `dsplit()`: 깊이 분할, 3차원으로 분할

```
a3 = np.arange(1, 28).reshape(3, 3, 3)
print(a3)
b3, c3 = np.dsplit(a3, [2])
print(b3)
print(c3)
```

## 배열 연산

### 브로드캐스팅(Broadcasting)

```
a1 = np.array([1,2,3])
print(a1)
print(a1+5)
```

```
a2 = np.arange(1,10).reshape(3,3)
print(a1)
print(a1+a2)
```

### 산술 연산(Arithmetic Operators)

```
a1 = np.arange(1,10)
print(a1)
print(a1+1)
print(np.add(a1, 10))
print(a1-2)
print(np.subtract(a1, 10))
print(-a1) #negative 붙이면 다 음수로 바뀐다
print(np.negative(a1))
print(a1 * 2)
print(np.multiply(a1, 2))
print(a1/2)
print(np.divide(a1, 2))
print(a1//2)
print(np.floor_divide(a1, 2)) #floor_divide = 나눗셈 내림
print(a1 ** 2)
print(np.power(a1, 2)) #power
print(a1 % 2)
print(np.mod(a1, 2)) #나머지
```

```
a1 = np.arange(1,10)
print(a1)
b1 = np.random.randint(1,10,size=9)
print(b1)
print(a1 + b1)
print(a1 - b1)
```
